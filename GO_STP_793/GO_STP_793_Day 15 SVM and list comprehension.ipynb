```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.feature_names
```

```python
cancer.target_names
```

```python
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
df.head()
```

```python
df['target'] = cancer.target
```

```python
df
```

```python
df.shape
```

```python
# apply() - pandas function which can apply a given function/value to all or specific rows of the dataframe
df['cancer class'] = df.target.apply(lambda x: cancer.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['target', 'cancer class'], axis = 'columns')

Y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
print(X_train.shape)
print()
print(X_test.shape)
```

```python
from sklearn.svm import SVC

model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sc = []
for i in range(2, 10):
    print(i, end = " ")
    model_c = SVC(C = i)
    model_c.fit(X_train, y_train)
    sc.append(model.score(X_test, y_test))
    print(sc[i-2])
    
x = range(2, 10)
plt.plot(x, sc)
plt.show()
```
