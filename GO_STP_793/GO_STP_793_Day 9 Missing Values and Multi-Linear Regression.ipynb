```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student_data.csv')
df
```

```python
#fillna methods returns a dataframe with the nullvalues resolved
# id=f we want to change the dataframe we can set the 'inplace' 
# argument to 'True'

df = df.fillna(0)    #fill all null values with 0 or argument
df
```

```python
#fill null values accoring to column

new_df = df.fillna(
{
    'CS':0,
    'Math':0,
    'Name':'No_name'
})
new_df
```

```python
#method can be ffill, bfill, pad, none i.e. forward fill or backward
#it reads the data from the specified direction and fills data
# according to the last value read

new_df1 = df.fillna(method = 'bfill')
new_df1
```

```python
#here last column value will be copied
new_df2 = df.fillna(method = 'ffill', axis = 'columns')
new_df2
```

```python
#fills only the next 'limit' values and remaining NA values will be as
# is
new_df3 = df.fillna(method ='ffill', limit = 1)
new_df3
```

```python
#interpolate is mathematical function
new_df4 = df.interpolate()
new_df4
```

```python
#dropna() - any row with a na entry will be completely removed

new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')  #drop rows only of all values are null in a row
new_df
```

```python
new_df = df.dropna(thresh = 2)  # there must be data in two columns
# less than that and the row is deleted
new_df
```

```python
#replace(arg1, arg2) - replace agr1 with arg2
new_df = df.replace(np.NaN, 9999)
new_df
```

```python
new_df = df.replace(
    {
        'CS': 999, 
        'Math': 0, 
        'Name': 'Nothing'
    }
)
new_df
```

# Boston House Price Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston)
```

```python
boston.keys()
```

```python
boston['data'].shape
```

```python
bs = pd.DataFrame(boston.data, columns = boston.feature_names)
bs.head()
```

```python
bs['Price'] = boston.target
bs.head()
```

```python
y = bs['Price']
x = bs.drop('Price', axis = 1)
print(x.head())
print()
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.33, random_state = 5)
```

```python
lm = LinearRegression()
lm.fit(X_train, Y_train)
print(lm)
```

```python
Y_pred_train = lm.predict(X_train)
Y_pred_test = lm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(Y_test, Y_pred_test)
print(mse)
```
