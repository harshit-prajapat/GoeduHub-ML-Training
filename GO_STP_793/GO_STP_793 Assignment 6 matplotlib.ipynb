```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed
#represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as 
#car speed and put title Car Speed Measurement.


Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]

plt.plot(Days, Speed)
plt.title("Car Speed measurement")
plt.xlabel("days")
plt.ylabel("car speed")
plt.show()
```

```python
#2. Now to above car data apply some string formats like line style example green dotted line, marker shape 
#like +, change markersize, markerface color etc.

Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]

plt.subplot(221)
plt.plot(Days, Speed, 'g:')
plt.title("Car Speed measurement")
plt.xlabel("days")
plt.ylabel("car speed")

plt.subplot(222)
plt.plot(Days, Speed, 'b+--')
plt.title("Car Speed measurement")
plt.xlabel("days")
plt.ylabel("car speed")

plt.subplot(223)
plt.plot(Days, Speed, 'bo--', markersize=10, markerfacecolor='black')
plt.title("Car Speed measurement")
plt.xlabel("days")
plt.ylabel("car speed")

plt.subplot(224)
plt.plot(Days, Speed, linewidth=10)
plt.title("Car Speed measurement")
plt.xlabel("days")
plt.ylabel("car speed")

plt.show()
```

```python
#3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]

plt.title('Car Travel data')
plt.plot(days, max_speed, label = 'max_speed')
plt.plot(days, min_speed, label = 'min_speed')
plt.plot(days, avg_speed, label = 'avg_speed')
plt.xlabel('days')
plt.ylabel('speeds')
plt.grid()
plt.legend()
plt.show()
```

```python
#4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

angles = np.arange(0, 6.4, 0.1)
sine = np.sin(angles)

plt.xlabel('angles')
plt.ylabel('sin/cos')
plt.plot(angles, sine, label = 'sine angles')
plt.plot(angles, np.cos(angles), label = 'cosine angle')
plt.legend()
plt.show()
```

```python
#5. Plot Simple bar chart showing popularity of Programming Languages.
# Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
# Popularity = [56, 39, 34, 34, 29]
# Security = [44 ,36 ,55, 50, 42]
# Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using 
# barh function.

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]
security = [44 ,36 ,55, 50, 42]

# plt.subplot(121)
plt.bar(languages, popularity, width = 0.3, label = 'popularity', align='edge')
plt.bar(languages, security, width = 0.3, label = 'security')
plt.legend()
plt.show()

# plt.subplot(122)
plt.barh(languages, popularity, height=0.3, label = 'popularity', align='edge')
plt.barh(languages, security, height=0.3, label = 'security')
plt.legend()
plt.show()
```
