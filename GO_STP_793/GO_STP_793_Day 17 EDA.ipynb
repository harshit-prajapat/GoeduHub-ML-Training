# Bivariate Analysis

```python
sns.set_style('whitegrid')

sns.FacetGrid(data, hue = 'status').map(plt.scatter, 'age', 'year').add_legend()
```

```python
sns.set_style('whitegrid')

sns.FacetGrid(data, hue = 'status').map(plt.scatter, 'age', 'nodes').add_legend()
```

```python
sns.set_style('darkgrid')
sns.pairplot(data, hue='status', vars=['age', 'year', 'nodes'], palette = 'gist_rainbow')
```

# Univariate



```python
live = data.loc[data["status"] == 1]
dead = data.loc[data["status"] == 2]

plt.plot(live["age"], np.zeros_like(live['age']), 'o')
plt.plot(dead['age'], np.zeros_like(dead['age']), 'o')

plt.show()
```

```python
sns.FacetGrid(data, hue = 'status').map(sns.distplot, "year").add_legend()
```

```python
sns.distplot(data['age'], color='darkred')
```

```python
sns.histplot(data['age'])
```

```python
sns.boxplot(x = 'status', y = 'age', data = data)
plt.show()
```

```python
sns.boxplot(x = 'status', y = 'nodes', data = data)
plt.show()
#this shows that the people having less number of nodes are having 
#more chances to survive instead of those having more number of
#nodes.
```
