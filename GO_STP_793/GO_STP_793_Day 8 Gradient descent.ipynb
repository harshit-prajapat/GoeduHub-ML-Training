```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = 0
b = 0
rate = 0.001
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 7, 8])
```

```python
import math
def gradient(x, y):
    a = 0
    b = 0
    rate = 0.001
    n = len(x)
    old_cost = 0
    for i in range(1000):
        y_pred = a*x + b
        cost = (1/n) * sum([value**2 for value in (y - y_pred)])
#         plt.plot(x, y_pred)
        da = -(2/n) * sum(x*(y - y_pred))
        db = -(2/n) * sum(y - y_pred)
        a = a - rate*db
        b = b - rate*da
        
        if math.isclose(cost, old_cost):
            break
    old_cost = cost
    return a, b
```

```python
df = pd.read_csv("salaryData.csv")
df.head()
```

```python
x = np.array(df['YearsExperience'])
y = np.array(df['Salary'])

m, b = gradient(x, y)

print(m , b)
```
